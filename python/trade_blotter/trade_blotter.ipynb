```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('SampleInput.csv')
df
```

```python
df_end_output = pd.read_csv('SampleOutput.csv')
df_end_output
```

```python
df['SymbolBoughtHelper'] = 0
df.loc[df['Side'] == 'b', 'SymbolBoughtHelper'] = 1

# SymbolBought
df['SymbolBought'] = df['FillSize'].mul(df['SymbolBoughtHelper']).groupby(df['Symbol']).cumsum()
```

```python
df['SymbolSoldHelper'] = 0
df.loc[df['Side'] != 'b', 'SymbolSoldHelper'] = 1

# SymbolSold
df['SymbolSold'] = df['FillSize'].mul(df['SymbolSoldHelper']).groupby(df['Symbol']).cumsum()
```

```python
df['SymbolPositionHelper'] = -1
df.loc[df['Side'] == 'b', 'SymbolPositionHelper'] = 1

# SymbolPosition
df['SymbolPosition'] = df['FillSize'].mul(df['SymbolPositionHelper']).groupby(df['Symbol']).cumsum()
```

```python
# SymbolNotional
df['SymbolNotional'] = df['FillSize'] * df['FillPrice']
```

```python
df['ExchangeBoughtHelper'] = 0
df.loc[df['Side'] == 'b', 'ExchangeBoughtHelper'] = 1

# ExchangeBought
df['ExchangeBought'] = df['FillSize'].mul(df['ExchangeBoughtHelper']).groupby(df['FillExchange']).cumsum()
```

```python
df['ExchangeSoldHelper'] = 0
df.loc[df['Side'] != 'b', 'ExchangeSoldHelper'] = 1

# ExchangeSold
df['ExchangeSold'] = df['FillSize'].mul(df['ExchangeSoldHelper']).groupby(df['FillExchange']).cumsum()
```

```python
df['TotalBoughtHelper'] = 0
df.loc[df['Side'] == 'b', 'TotalBoughtHelper'] = 1

# TotalBought
df['TotalBought'] = df['FillSize'].mul(df['TotalBoughtHelper']).cumsum()
```

```python
df['TotalSoldHelper'] = 0
df.loc[df['Side'] != 'b', 'TotalSoldHelper'] = 1

# TotalSold
df['TotalSold'] = df['FillSize'].mul(df['TotalSoldHelper']).cumsum()
```

```python
df['TotalBoughtNotionalHelper'] = 0
df.loc[df['Side'] == 'b', 'TotalBoughtNotionalHelper'] = 1

# TotalBought
df['TotalBoughtNotional'] = df['SymbolNotional'].mul(df['TotalBoughtNotionalHelper']).cumsum()
```

```python
df['TotalSoldNotionalHelper'] = 0
df.loc[df['Side'] != 'b', 'TotalSoldNotionalHelper'] = 1

# TotalSoldNotational
df['TotalSoldNotional'] = df['SymbolNotional'].mul(df['TotalSoldNotionalHelper']).cumsum()
```

```python
df.drop(['SymbolBoughtHelper', 'SymbolSoldHelper', 'SymbolPositionHelper', 'ExchangeBoughtHelper',
         'ExchangeSoldHelper', 'TotalBoughtHelper', 'TotalSoldHelper', 'TotalBoughtNotionalHelper',
         'TotalSoldNotionalHelper'], axis = 1, inplace = True)

df
```
