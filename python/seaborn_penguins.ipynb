```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sns.set_style('darkgrid')
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
sns.displot(penguins['body_mass_g'], bins = 5, kde = True, rug = True)
```

```python
sns.kdeplot(penguins['body_mass_g'])
```

```python
sns.rugplot(penguins['body_mass_g'])
```

```python
sns.jointplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = penguins,
             kind = 'hex')
```

```python
sns.pairplot(penguins, hue = 'island')
```

```python
sns.barplot(x = 'sex', y = 'body_mass_g', data = penguins,
           estimator = np.mean, hue = 'island')
```

```python
sns.countplot(x = 'sex', data = penguins, hue = 'species')
```

```python
sns.boxplot(x = 'island', y = 'flipper_length_mm', data = penguins,
            hue = 'sex')
```

```python
sns.violinplot(x = 'island', y = 'flipper_length_mm', data = penguins,
              hue = 'sex', split = True)
```

```python
sns.stripplot(x = 'island', y = 'flipper_length_mm', data = penguins,
             hue = 'sex', dodge = True, jitter = False)
```

```python
sns.swarmplot(x = 'island', y = 'flipper_length_mm', data = penguins)
```

```python
sns.catplot(x = 'island', y = 'flipper_length_mm', data = penguins,
            kind = 'bar')
```

```python
sns.pairplot(penguins)
```

```python
g = sns.PairGrid(penguins)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data = penguins, col = 'island', row = 'sex')
g.map(plt.scatter, 'body_mass_g', 'flipper_length_mm')
```

```python
sns.lmplot(x = 'body_mass_g', y = 'flipper_length_mm', data = penguins,
          col = 'sex', row = 'island', hue = 'species',
          aspect = 2, height = 3)
```
