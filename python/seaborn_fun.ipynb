```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sns.set_style('darkgrid')
```

```python
cars = sns.load_dataset('mpg')
```

```python
cars.head()
```

```python
cars.corr()
```

```python
sns.heatmap(cars.corr(), annot = True, cmap = 'coolwarm')
```

```python
cars_pivot = cars.pivot_table(index = 'model_year', columns = 'origin',
                             values = 'mpg')
```

```python
cars_pivot
```

```python
sns.heatmap(cars_pivot, annot = True, cmap = 'plasma', linewidth = 1)
```

```python
sns.clustermap(cars_pivot, cmap = 'magma',
               linewidth = 1, standard_scale = 1)
```
